```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip", data=tips)
plt.title("scatter plot of total bill vs tip")
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.barplot(x="day",="total_bill", data=tips)
plt.title("bar plot of total bill vs tip")
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill", data=tips)
plt.title("boxplot of total bill vs tip")
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.violinplot(x="day",y="total_bill", data=tips)
plt.title("violin plot of total bill vs tip")
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
correlation_matrix = iris.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation heatmap of Iris Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
correlation_matrix = iris.corr(numeric_only=True)
sns.pairplot(iris)
plt.title("Pair plot of Iris Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
plt.title(" Jointplot of of tips Dataset")
plt.xlabel("Total Bill($)")
plt.ylabel("Tips($)")
plt.show()
```
